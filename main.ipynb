```python
train = pd.read_csv("/train.csv")
train.head()
```

```python
# Memeriksa tipe data
train.info()
```

```python
# Menampilkan statistik deskriptif dari datasets
train.describe(include="all")
```

```python
# Memeriksa jumlah nilai yang hilang di setiap kolom / missing value
missing_values = train.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Mengataai missing values
# Pertama pisahkan kolom yang memiliki missing values lebih dari 75% dan kurang dari 75%
less = missing_values[missing_values < 1000].index
over = missing_values[missing_values >= 1000].index
```

```python
# Mengisi nilai yang hilang dengan median untuk kolom numerik
# Memilih nama-nama kolom dari dataframe yang memiliki tipe data numerik dari subset kolom yang ditentukan oleh less
numeric_features = train[less].select_dtypes(include=["number"]).columns
# Mengisi semua nilai yang hilang (NaN) pada kolom-kolom numerik tersebut dengan nilai median dari masing-masing kolom
train[numeric_features] = train[numeric_features].fillna(train[numeric_features].median())
```

```python
# Mengisi nilai yang hilang dengan modus untuk kolom kategori
kategorical_features = train[less].select_dtypes(include=["object"]).columns

for columns in kategorical_features:
    train[columns] = train[columns].fillna(train[columns].mode()[0])
```

```python
# Menghapus kolom dengan terlalu banyak nilai yang hilang
df = train.drop(columns=over)
```

```python
# Melakukan pemeriksaan terhadap data yang sudah melewati tahapan verifikasi missing value
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Memeriksa apakah dataset memiliki outlier atau tidak
for features in numeric_features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[features])
    plt.title(f"Box plot of {features}")
    plt.show()
```

```python
# Mengidentifikasi outliers menggunakan IQR
Q1 = df[numeric_features].quantile(0.25)
Q3 = df[numeric_features].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Menghapus outlier berdasarkan perhitungan di atas
# Filter dataframe untuk ganya menyimpan baris yang tidak mengandung outliers pada kolom numerik
condition = ~((df[numeric_features] < (Q1 - 1.5 * IQR)) | (df[numeric_features] > (Q3 + 1.5 * IQR))).any(axis=1)
df_filtered_numeric = df.loc[condition, numeric_features]

# Menggabungkan kembali dengan kolom kategorikal
kategorical_features = df.select_dtypes(include=["object"]).columns
df = pd.concat([df_filtered_numeric, df.loc[condition, kategorical_features]], axis=1)
```

```python
for features in numeric_features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[features])
    plt.title(f"Box plot of {features}")
    plt.show()
```

```python
# Melakukan standarisasi data
scaler = StandardScaler()
df[numeric_features] = scaler.fit_transform(df[numeric_features])
```

```python
# Memisahkan fitur (X) dan target (Y)
x = df_lencoder.drop(columns=["SalePrice"])
y = df_lencoder["SalePrice"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Membagi dataset menjadi training dan testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
# Menghitung panjang/jumlah data
print("Jumlah data: ", len(x))
# Menghitung panjang/jumlah data pada x_train
print("Jumlah data latih: ", len(x_train))
# Menghitung panjang/jumlah data pada x_test
print("Jumlah data test: ", len(x_test))
```

```python
# Melatih model 1 dengan algoritma Least Angle Regression
from sklearn import linear_model
lars = linear_model.Lars(n_nonzero_coefs=1).fit(x_train, y_train)
```

```python
# Melatih model 2 dengan algoritma Linear Regression
from sklearn.linear_model import LinearRegression
LR = LinearRegression().fit(x_train, y_train)
```

```python
# Melatih model 3 dengan algoritma Gradient Boosting Regressor
from sklearn.ensemble import GradientBoostingRegressor
GBR = GradientBoostingRegressor(random_state=184)
GBR.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Evaluasi pada model LARS
pred_lars = lars.predict(x_test)
mae_lars = mean_absolute_error(y_test, pred_lars)
mse_lars = mean_squared_error(y_test, pred_lars)
r2_lars = r2_score(y_test, pred_lars)

# Membuat dictionary untuk menyimpan hasil evaluasi
data = {
    "MAE": [mae_lars],
    "MSE": [mse_lars],
    'R2': [r2_lars]
}

# Konversi  dictionary menjadi Dataframe
df_result = pd.DataFrame(data, index=["Lars"])
df_result
```

```python
# Evaluasi pada model Linear Regression
pred_LR = LR.predict(x_test)
mae_LR = mean_absolute_error(y_test, pred_LR)
mse_LR = mean_squared_error(y_test, pred_LR)
r2_LR = r2_score(y_test, pred_LR)   

# Menambahkan hasil evaluasi LR ke Dataframe
df_result.loc["Linear Regression"] = [mae_LR, mse_LR, r2_LR]
df_result
```

```python
# Evaluasi pada model Gradient Boosting Regressor
pred_GBR = GBR.predict(x_test)
mae_GBR = mean_absolute_error(y_test, pred_GBR)
mse_GBR = mean_squared_error(y_test, pred_GBR)
r2_GBR = r2_score(y_test, pred_GBR)

# Menambahkan hasik evaluasi GBR ke Dataframe
df_result.loc["GradientBoostingRegressor"] = [mae_GBR, mse_GBR, r2_GBR]
df_result
```
